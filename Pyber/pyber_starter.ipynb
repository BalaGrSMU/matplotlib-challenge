```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(ride_data,city_data,on='city',how = 'left')
# Display the data table for preview
combined_data.head()
```

```python
# Create separate dataframes for urban, rural and suburban type
urban_type = combined_data[combined_data["type"] == "Urban"].groupby([combined_data["city"]])
rural_type = combined_data[combined_data["type"] == "Rural"].groupby([combined_data["city"]])
suburban_type = combined_data[combined_data["type"] == "Suburban"].groupby([combined_data["city"]])

# Urban calculation of rides per city, average fare, bubble size using driver count
x_urban = urban_type['ride_id'].count()
y_urban = urban_type['fare'].mean()
s_urban = urban_type['driver_count'].mean()

#Rural calculation of rides per city, average fare, bubble size using driver count
x_rural = rural_type['ride_id'].count()
y_rural = rural_type['fare'].mean()
s_rural = rural_type['driver_count'].mean()

# Suburban calculation of rides per city, average fare, bubble size using driver count
x_suburban = suburban_type['ride_id'].count()
y_suburban = suburban_type['fare'].mean()
s_suburban = suburban_type['driver_count'].mean()
```

```python
# create scatter plots for each city type
plt.scatter(x_rural, y_rural, s=s_rural*10,label = "Rural", alpha = 0.75, color =["blue"],edgecolor ="black", marker ="o")
plt.scatter(x_urban, y_urban, s=s_urban*10,label = "Urban", alpha=0.75, color=["coral"], edgecolor="black", marker="o")
plt.scatter(x_suburban, y_suburban, label = "Suburban", s=s_suburban*10,alpha = 0.75, color =["gold"],edgecolor ="black", marker ="o")


# create title, xlabe, ylabel
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides per city")
plt.ylabel("Average Fare ($)")

# Create a legend
legend = plt.legend(fontsize = 8, title= "City Types", loc="best") 

# set limits for x-axis and y-axis
plt.xlim(0,45)
plt.grid()

# Adjust the legend 
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(50,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Save figure with the parameter bbox_inches to fit the whole image
plt.savefig("images/PyberRideSharingData.png", bbox_inches="tight")

# Show plot
plt.show()
```

```python
# create pie chart essentials
pies = ["Rural", "Suburban", "Urban"]
type_grouping = combined_data.groupby(["type"])
pie_group = type_grouping["fare"].sum()
colors = ["gold","coral","green"]
explode = (0,0,0.1)

# create a pie chart based upon the above data
plt.pie(pie_group, explode=explode, labels=pies, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")

# plot title
plt.title("% of Total Fares by City Type")

# Save an image of our chart and print the final product to the screen
plt.savefig("images/PyPiescitytype.png")
plt.show()
```
